```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate


%matplotlib inline
```

```python
def quadratic_func(x, c=-10, a=90):
    return 8*(x**4) + 2*(x**3)+a*(x**2) -2*x +c
```

```python
def cos_func(x):
    return 10*(np.cos(5*x))
```

```python
def result_func(x, c=-10, a=90):
    return cos_func(x) - quadratic_func(x, c=c, a=a)
```

```python
x_points = np.linspace(-20, 5 * np.pi, 100000)  # 100000 точек
```

##  1)Ответ на первый вопрос: построение графиков

```python
y1_of_quadratic_func = [quadratic_func(x) for x in x_points]
y2_of_cos_func = [cos_func(x) for x in x_points]

y1_of_quadratic_func = np.array(y1_of_quadratic_func)
y2_of_cos_func = np.array(y2_of_cos_func)

plt.figure(figsize=(20,10))

plt.xlim(-1, 12)
plt.ylim(-12, 12)
plt.plot(x_points, y1_of_quadratic_func, label='y=8x^4+2x^3+90x^2-2x-10')
plt.plot(x_points, y2_of_cos_func, label='y=10*cos(5x)')
plt.grid(True)

idx = np.argwhere(np.diff(np.sign(y1_of_quadratic_func - y2_of_cos_func))).flatten()
plt.plot(x_points[idx], y1_of_quadratic_func[idx], 'ro')

print(f'Точка пересечения графиков: x={x_points[idx]}, y={y1_of_quadratic_func[idx]}')


plt.title('График 1')
plt.ylabel('Y(x)')
plt.xlabel('x')
plt.legend()
plt.show()
```

```python
c_points_where_area_zero = []
```

```python
def get_depend_on_c_calculate(c, a=90):
    x_points = np.linspace(-20, 5 * np.pi, 100000)  # 100000 точек
    
    y1_of_quadratic_func = [quadratic_func(x, c=c, a=a) for x in x_points]
    y2_of_cos_func = [cos_func(x) for x in x_points]


    y1_of_quadratic_func = np.array(y1_of_quadratic_func)
    y2_of_cos_func = np.array(y2_of_cos_func)

    
    
    idx = np.argwhere(np.diff(np.sign(y1_of_quadratic_func - y2_of_cos_func))).flatten()
    # print(idx) crossed points
    if not idx.any():
        c_points_where_area_zero.append(c)
        return 0
 



    result_area_cross_2_graphs_local = scipy.integrate.quad(result_func, x_points[idx][0], x_points[idx][1])

    return result_area_cross_2_graphs_local[0]
    
```

```python
def get_depend_s_on_c():
    c_points = [i for i in range(-10, 15)]
    result_area = []
    for point in c_points:
        result_area.append(get_depend_on_c_calculate(c=point))
        
    return result_area, c_points
```

```python
result_are_get_s_from_only_c, result_x_points_task_2 = get_depend_s_on_c()
```

```python
def get_depend_s_on_c_and_a():
    c_points = [i for i in range(-10, 80)]
    a_points = [i for i in range(10, 100)]
    
    result_area = []
    for i in range(0, 20):
        c = c_points[i]
        a = a_points[i]
        result_area.append(get_depend_on_c_calculate(c, a=a))

        
    return result_area, c_points, a_points
```
